```python
df2
```

```python
df2.loc[1,'quiz2'] = round(df2['quiz2'].mean(),1)

df2.loc[7,'quiz1'] = round(df2['quiz1'].mean(),1)
df2.loc[2,'assignment'] = round(df2['assignment'].mean(),1)
df2.loc[4,'quiz1'] = 10.0
df2.loc[4,'attendance'] = 100
df2
```

```python
df2['total_score'] = df2['quiz1']+df2['quiz2']+df2['assignment']+df2['final']+df2['midterm']

df2
```

```python
df2['grade'] = pd.cut(df2['total_score'],
                              bins =[0,20,40,60,80,100,120],
                              labels =['F','E','D','C','B','A'])
df2
```

```python
import numpy as np

df2['result'] = np.where(df2['total_score'] > 60, 'Pass', 'Fail')
df2
```

```python
df2.groupby('course')['total_score'].mean()
```

```python
df2.groupby('grade')['total_score'].count()
```

```python
(df2['result'] == 'Fail').sum()
```
